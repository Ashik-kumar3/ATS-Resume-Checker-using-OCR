```python
import cv2
import pytesseract
import re
```

```python
resume=cv2.imread('res.jpg')
cv2.imshow("frame",resume)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
pytesseract.pytesseract.tesseract_cmd=r"C:\Program Files\Tesseract-OCR\tesseract.exe"
text = pytesseract.image_to_string(resume)
print(text)
```

```python
dict = pytesseract.image_to_data(resume,output_type=pytesseract.Output.DICT)
dict
```

```python
n = len(dict['text'])
n
```

```python
pattern = r'^[\sA-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
for i in range(n):
    if re.search(pattern,dict['text'][i]):
        print('Email : ',dict['text'][i])
#         x = dict['left'][i]
#         y = dict['top'][i]
#         w = dict['width'][i]
#         h = dict['height'][i]
#         text1 = dict['text'][i]
#         cv2.rectangle(resume,(x,y),(x+w,y+h),(255,0,0),2)
#         cv2.putText(resume,text1,(x-1,y-1),fontFace=cv2.FONT_HERSHEY_PLAIN,fontScale=1,color=(255,0,0))
# cv2.imshow('frame',resume)
# cv2.waitKey(5000)
# cv2.destroyAllWindows()
```

```python
pattern =r'[0-9]{10}'
for i in range(n):
    if re.search(pattern,dict['text'][i]):
        print('Phone No : ',dict['text'][i])
```

```python
JD = ['Machine','Learning','Deep','Learning','Generative','AI','computer','vision','PyTorch','TensorFlow','Python','Pandas',' SQL','PowerBI','Excel','CNN','GAN']
jd_len = len(JD)
count = 0
for i in range(n):
    if dict['text'][i] in JD :
        x = dict['left'][i]
        y = dict['top'][i]
        w = dict['width'][i]
        h = dict['height'][i]
        text2 = dict['text'][i]
        cv2.rectangle(resume,(x,y),(x+w,y+h),(255,0,0),2)
        JD.remove(dict['text'][i])
        count+=1
cv2.imshow('resume',resume)
cv2.waitKey(0)
cv2.destroyAllWindows()


percentage = (count/jd_len)*100
print('Percentage :',percentage)
```
